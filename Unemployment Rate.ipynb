```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("unemployment analysis.csv")
df.head()
```

## Check our data before we begin to analyze

```python
df.info()
```

```python
df.describe()
```

## Checking to see if we have any gap in our Dataframe

```python
df.isna().sum()
```

```python
df.set_index("Country Name", inplace = True)
```

## Everything looks good, perfect!, now let's analyze!

```python
df_t = df.T
df_t.head()
```

```python
df_deu = df_t["Germany"]
df_deu
```

```python
fig, ax = plt.subplots(1,1,figsize=(12, 6))

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

plt.plot(df_deu.iloc[1:])

plt.title("Unemployment Rate in Germany")
plt.xticks(rotation = 60)
plt.xlabel("Years")
plt.ylabel("Unemployment Rate")
plt.grid(axis = "both", ls = "--")

plt.show()
```

```python
df_eu = df_t[["Germany","Belgium","France","Netherlands","Spain","Italy"]]
df_eu[1:].head()
```

## Now let's compare Germany with some countries in the EU

```python
fig, ax = plt.subplots(1,1,figsize=(12, 6))

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

plt.plot(df_eu["Germany"].iloc[1:], label = "Germany")
plt.plot(df_eu["Belgium"].iloc[1:], label = "Belgium")
plt.plot(df_eu["France"].iloc[1:], label = "France")
plt.plot(df_eu["Netherlands"].iloc[1:], label = "Netherlands")
plt.plot(df_eu["Spain"].iloc[1:], label = "Spain")
plt.plot(df_eu["Italy"].iloc[1:], label = "Italy")

plt.title("Unemployment Rate European Countries")
plt.xticks(rotation = 60)
plt.xlabel("Years")
plt.ylabel("Unemployment Rate")
plt.grid(axis = "both", ls = "--")

plt.legend(loc = 1)

plt.show()
```

## Now let's take a look at the countries with the lowest unemployment rate

```python
df_low = df.sort_values(by = "2021")
df_low.head()
```

```python
df_low.T.iloc[1:,:3].head()
```

## and the countries with the highest Unemployment Rate!

```python
df_hi = df.sort_values(by = "2021", ascending = False)
df_hi.head()
```

```python
df_hi.T.iloc[1:,:3].head()
```

## let's see those countries in a Chart!

```python
fig, ax = plt.subplots(1,1,figsize=(12, 6))

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)
    
plt.plot(df_low.T["Qatar"].iloc[1:], label = "Qatar")
plt.plot(df_low.T["Cambodia"].iloc[1:], label = "Cambodia")
plt.plot(df_low.T["Niger"].iloc[1:], label = "Niger")
plt.plot(df_low.T["South Africa"].iloc[1:], label = "South Africa")
plt.plot(df_low.T["Djibouti"].iloc[1:], label = "Djibouti")
plt.plot(df_low.T["Eswatini"].iloc[1:], label = "Eswatini")
    
plt.title("Countries with the highest and lowest Unemployment Rate")
plt.xticks(rotation = 60)
plt.xlabel("Years")
plt.ylabel("Unemployment Rate")
plt.grid(axis = "both", ls = "--")

plt.legend(loc = 5)

plt.show()
```
